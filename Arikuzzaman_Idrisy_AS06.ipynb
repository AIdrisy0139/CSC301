```python
import math

stepSize = 0
# initVel = 0
# initPos = -math.pi/4
g = 32.17
def sin(x):
    return math.sin(x)

def Force(pos):
    #print("Force Function")
    #print(g)
    return -g*sin(pos)

def nextVelocity(vT,xT):
    return vT+Force(xT)*(stepSize/2)

def nextPosition(xT,vT):
    return xT + vT*stepSize

def isCriticalTime(currPos,nextPos):
    if(currPos<0 and nextPos > 0):
        return True
    else:
        return False

def getCritTimes(a,atArray):
    myArray = []
    for i in range(0,(len(a)-1)):
        if(isCriticalTime(a[i],a[i+1])):
            myArray.append(atArray[i] + (stepSize/2)) 
    return myArray
            
def arrTupleDiffAvg(a):
    ssf =0
    for i in range(0,(len(a)-1)):
        ssf = ssf + abs((a[i+1]-a[i]))
    return ssf/math.ceil(len(a)/2)

# Returns average period time of periods in  between zeroTime and totalTime
def verlet(zeroPos,zeroVel,zeroTime,timeStep,totalTime):

    critTimeArray = [] 
    currentTime = zeroTime
    allTimeArr = [currentTime]
    position = zeroPos
    posArray = [position]
    velocity = zeroVel
    velArray=[velocity]
    
    global stepSize
    stepSize = timeStep
    while(currentTime<totalTime):
        # print("while")
        currentTime = currentTime + stepSize 
        allTimeArr.append(currentTime)
        halfNextVel = nextVelocity(zeroVel,position)
        nextPos = nextPosition(position,halfNextVel)
        posArray.append(nextPos)
        fullNextVel = nextVelocity(halfNextVel,nextPos)
        velArray.append(fullNextVel)

        position = nextPos
        velocity =fullNextVel
    critTimeArray = getCritTimes(posArray,allTimeArr)
    # print("Current Time: " + str(currentTime))
    # print("X")
    # print(critTimeArray)
    return arrTupleDiffAvg(critTimeArray)


def runVerlet():
    assignedPos = -math.pi/4
    assignedVel = 0
    startTime = 0
    step=1
    run=200
    print("--------Increasing Length of Run Time---------")
    print("Time Step Size: {} seconds".format(step))
    for runTime in range(200,1100,100):
       print("Average Period for {} seconds = {}".format(runTime , verlet(assignedPos,assignedVel,startTime,step,runTime)))
    print("--------Decreasing Step Size ---------------")
    print("Runtime = " + str(run))
    stepArray = [4,3,2,1,.75,.5,.25]
    for stepSize in stepArray:
        print("Step Size = {} | Average Period = {}".format(stepSize,verlet(assignedPos,assignedVel,startTime,stepSize,run)))

runVerlet()
```

# Analysis

## Increasing Length of Run Time

This exhibits some weird behavior. As the runtime increases the period decreases as. This shouldn't happen because we're in an Inertial Reference frame with no dampening forces(i.e. friction or air drag). The output data would make it seem as if the Pendulum is being driven, as the period is getting shorter over time. It does seem to stablize as the run time approaches 1000. The magnitude of the accerlation decreases, but its still there as we see changes in the last few digits.

## Decreasing Step Size
This also exhbiits unexpected behavior, as the period decreases with the step size. When printing out the critical time array, it is seen that the array is not equal for all runs as would be expected based on the physics of the problem. My hypothesis is that this anomoly occurs because when the step size changes our corresponding critical positions are different, which means they occur at different times. Lower step size in time leads to lower increments in poisition as well, therefore we ultimately have different critical points in each test.
