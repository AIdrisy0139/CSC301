# ANALYSIS

## Sin Approximations
    
    As expected as n increases regardless of the value of x the error between the 'exact' and estimated values get closer. This makes sense as the taylor polynomial for sin is a alternating series, so as n increases the approximation gets closer to the 'exact' value of sin(x). As x increases however the average of the errors gets worse. When x =1 the error underflows by n=4 with a lower average where as by the time x=4 the error never underflows. Beyond this there is not much to say besides that the Taylor Polynomial of Sin is a very accurate approximation given that all you really need are the values of sin from [0, pi].
    
## Cos Approximations
    Generaly speaking the analysis for the estimation of sin holds for the estimation of x. There are some saliant differences. The error does not increase porpotionaly with x as it did for sin. My hypothesis would be because sin(0)=0 and increasing positively as x approaches pi/2 where as cos(0)=1 and decreases, eventually crossing the x axis at pi/4. An increase of n still makes the function more accurate and the taylor approximation is still serviceably close.
    
## Natural Exponent Approximations
    The taylor approximation for exp follows the trends noticed for sin and cos, getting more accurate with a higher n and the errors getting worse with a higher x. The absolute errors for exp get pretty bad however the relative error stays at a acceptable range. This makes sense as exp grows very fast, getting very big, but since the value is so big the error is much more negligible.
    
## Natural Log Approximations
    The same dependance on x and n are seen in the ln approximations. The rate of growth in the errors however is very slow, wich correlates to ln's own slow growth.
    
## Summary
    The higher the n and the lower the x the better approximaton will be to the exact value. The growth in the error of the approximations as x changes seems to be dependant on the first derivative of said function.
